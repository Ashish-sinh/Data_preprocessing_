# Discretizer  || bining

```python
# ## KBinDiscretizer () : 
#                         'this type of data conver the data into giver 
#     number of bins using the k near clustring. : 
        
# 'n_sample' :  'encode' = 'onehot','ordinal','onehot-dense' 
# strategy : 'quntile','uniform','k means' 

# genrally use quntile strategy and ordinal encode
```

```python
from sklearn.datasets import load_iris 
from sklearn.preprocessing import KBinsDiscretizer
x = pd.DataFrame(load_iris().data , columns  =load_iris().feature_names )

descritizer  = KBinsDiscretizer (n_bins = 5, encode = 'ordinal', strategy ='quantile' ) 
x_desc = descritizer.fit_transform(x)

df = pd.DataFrame( x_desc, columns = load_iris().feature_names )
df

import  matplotlib.pyplot as plt 

fig = plt.figure(figsize= (10,6)) 

plt.subplot(1,2,1)
x['sepal width (cm)'].hist(color = 'red')
plt.title( 'before') 

plt.subplot(1,2,2)
df['sepal width (cm)'].hist(color = 'blue')
plt.title('after')

plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures 

pf= PolynomialFeatures () 

x = np.arange(4).reshape(2,2)
x_new = pf.fit_transform(x)
print('old is \n ',x)
print('new is \n ',x_new)

# If the input feature matrix has 4 features and the degree is set to 2, then the output feature matrix would be:
# [1, a1, a2, a3, a4, a1^2, a1a2, a1a3, a1a4, a2^2, a2a3, a2a4, a3^2, a3a4, a4^2]
```

```python
from sklearn.preprocessing import PolynomialFeatures ,SplineTransformer 

x = np.linspace( -1, 10, 100 ).reshape(-1,1)
y = np.sin(x).reshape(-1,1)

poly = PolynomialFeatures () 
x_poly = poly.fit_transform(x)
y_poly = poly.fit_transform(y)

spline = SplineTransformer () 
x_spline = spline.fit_transform(x)
y_spline = spline.fit_transform(y)

import matplotlib.pyplot as plt

fig = plt.figure ( figsize = (12,5))
plt.subplot(1,3,1)
plt.plot(x,y,label ='original')
plt.legend() 
plt.title('original')

plt.subplot(1,3,2)
plt.plot(x_poly,y_poly,label = 'poly')
plt.legend()
plt.title('polynomial')


plt.subplot(1,3,3)
plt.plot(x_spline,y_spline ,label = 'spline')
plt.title('splinetransformer')


plt.legend() 
plt.show ()
```
